```python
#Importing the packages for the EDA Process
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore",category=DeprecationWarning)
```

```python
#Importing the dataset using the pandas dataframe
por = pd.read_csv("Downloads/student-por.csv")
mat = pd.read_csv("Downloads/student-mat.csv")
```

```python
#Checking for columns and null values for mat dataframe
mat.info()
```

```python
#Concating the mat and por dataframe for EDA process
data = pd.concat([mat,por])
data.head()
```

```python
#Manipulating the columns records in the concated data 
data['sex'] = data['sex'].map({'F':'Female','M':'Male'})
data['school'] = data['school'].map({'GP':'Gabriel Pereira','MS':'Mousinho da Silveira'})
data['Dalc'] = data['Dalc'].map({1:'Very Low',2:'Below Average',3:'Average',4:'Above Average',5:'Very High'})
data['Walc'] = data['Walc'].map({1:'Very Low',2:'Below Average',3:'Average',4:'Above Average',5:'Very High'})
data['goout'] = data['goout'].map({1:'Very Low',2:'Below Average',3:'Average',4:'Above Average',5:'Very High'})
data['Pstatus'] = data['Pstatus'].map({'A':'Apart','T':'Living Together'})
data['address'] = data['address'].map({'U':'Urban','R':'Rural'})
data['health'] = data['health'].map({1:'Very Bad',2:'Bad',3:'Moderate',4:'Good',5:'Very Good'})
data['famsize'] = data['famsize'].map({'GT3':'Greater Than 3','LE3':'Less Or Equal To 3'})
```

```python
#Remaning the columns of tha dataframe
data = data.rename(columns={'Dalc':'Daily Alcohol Consumption',
                            'Walc':'Weekly Alcohol Consumption',
                            'Pstatus':'Parents Cohabitation Status',
                            'goout':'Going Out With Friends',
                            'famsize':'Family Size',
                            'address':'Address'})
```

```python
#Viewing the top 5 data from the concated dataframe
data.head()
```

```python
#Checking the fields and the null values in tha dataframe
data.info()
```

```python
#To know the shape of the dataframe
data.shape
```

```python
#Visualize the alcohol consumption on the daily basis
plt.figure(figsize=(11,7))
sns.set(font_scale=1)
sns.set_style('whitegrid')
Daily = data.groupby('Daily Alcohol Consumption').size()
x = list(Daily)
y = ['Above Average','Average','Below Average','Very High','Very Low']
ax=sns.barplot(y, x,order=['Very Low','Below Average','Average','Above Average','Very High'],
               palette='Blues_d')
ax.bar_label(ax.containers[0])
plt.title("Alcohol Consumed By Syudents On Daily Basis")
plt.xlabel('Category')
plt.ylabel('Total Count')
plt.show()
```

```python
#Visualize the alcohol consumption on the weekly basis
plt.figure(figsize=(11,7))
sns.set(font_scale=1)
sns.set_style('whitegrid')
Weekly = data.groupby('Weekly Alcohol Consumption').size()
x = list(Weekly)
y = ['Above Average','Average','Below Average','Very High','Very Low']
ax=sns.barplot(y, x,order=['Very Low','Below Average','Average','Above Average','Very High'],palette='rocket')
ax.bar_label(ax.containers[0])
plt.title("Alcohol Consumed By Syudents On Weekly Basis")
plt.xlabel('Category')
plt.ylabel('Total Count')
plt.show()
```

```python
#Extracting the data based on the high consumption of alcohol by the student
df = data[data['Daily Alcohol Consumption']=='Very High']
df1 = data[data['Daily Alcohol Consumption']=='Above Average']
dfs = data[data['Weekly Alcohol Consumption']=='Very High']
dfs1 = data[data['Weekly Alcohol Consumption']=='Above Average']
df2 = pd.concat([df,df1,dfs,dfs1])
```

```python
#To know the shape of the dataframe
df2.shape
```

```python
#To know the null values and the columns in the dataframe df2
df2.info()
```

```python
#Extract the top5 records from tha dataframe df2
df2.head()
```

```python
# Visualize based on the distribution of the school who consumed alcohol
vb = df2.groupby('school').size()
plt.figure(figsize=(15,5))
plt.ylabel('School')
plt.xlabel('Count')
plt.title('Distribution Based On School')
vbb = list(vb)
cbb = [' Gabriel Pereira','Mousinho da Silveira']
gh = sns.barplot(vbb,cbb,palette=['olive','seagreen'])
sns.set_style('whitegrid')
gh.bar_label(gh.containers[0])
plt.show()
```

```python
# Visualize based on the distribution of the sex who consumed alcohol
df = data[data['Daily Alcohol Consumption']=='Very High']
df1 = data[data['Daily Alcohol Consumption']=='Above Average']
dfs = data[data['Weekly Alcohol Consumption']=='Very High']
dfs1 = data[data['Weekly Alcohol Consumption']=='Above Average']
df2 = pd.concat([df,df1,dfs,dfs1])
School = df2.groupby('sex').size()
plt.figure(figsize=(15,5))
ss = list(School)
zz = ['Male','Female']
sx=sns.barplot(ss,zz,order=['Female','Male'],palette=['darkgoldenrod','rosybrown'])
sns.set_style('whitegrid')
sx.bar_label(sx.containers[0])
plt.title('Distribution Based On Sex')
plt.xlabel('Total Count')
plt.ylabel('Sex')
plt.show()
```

```python
# Visualize based on the distribution of the age who consumed alcohol
xc = df2.groupby('age').size()
plt.figure(figsize=(15,9))
plt.xlabel('Total Count')
plt.ylabel('Age')
cc = list(xc)
vc = ['15','16','17','18','19','20','21','22']
lx=sns.barplot(vc,cc,order=['17','16','18','15','19','20','22','21'],palette="Paired")
sns.set_style('whitegrid')
lx.bar_label(lx.containers[0])
plt.title('Distribution Based On the Age')
plt.show()
```

```python
# Visualize based on the distribution of the area who consumed alcohol
tnm = df2.groupby('Address').size()
plt.figure(figsize=(15,5))
plt.title('Distribution Based On Area Of The Student')
plt.xlabel('Total Count')
plt.ylabel('Area Of The Students')
tmm = list(tnm)
td = ['Urban','Rural']
sns.set_style('whitegrid')
bm = sns.barplot(tmm,td,order=['Rural','Urban'],palette=['#ff5000','#685000'])
bm.bar_label(bm.containers[0])
plt.show()
```

```python
# Visualize based on the distribution of thev going out with friends and consumed alcohol
rty = df2.groupby('Going Out With Friends').size()
plt.figure(figsize=(15,9))
plt.title('Distribution Based On Going Out With Friends')
plt.xlabel('Total Count')
plt.ylabel('Category')
rtyy = list(rty)
rte = ['Above Average','Average','Below Average','Very High','Very Low']
kk = sns.barplot(rte,rtyy,order=['Very High','Above Average','Average','Below Average','Very Low'],
                 palette="magma",saturation=0.4)
kk.bar_label(kk.containers[0])
sns.set_style('white')
plt.show()
```

```python
# Visualize based on the distribution of the parents cohabitation status who consumed alcohol
ty = df2.groupby('Parents Cohabitation Status').size()
plt.figure(figsize=(15,5))
plt.title('Distribution Based On Parents Cohabitation Status')
plt.xlabel('Total Count')
plt.ylabel('Parents Cohabitation Status')
tyy = list(ty)
tyu = ['Living Together','Apart']
bn = sns.barplot(tyy,tyu,order=['Apart','Living Together'],palette=['teal','darkkhaki'])
sns.set_style('whitegrid')
bn.bar_label(bn.containers[0])
plt.show()
```
